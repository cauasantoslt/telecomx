```python
df.info()
```

```python
valores_ausentes = df.isnull().sum()
valores_ausentes
```

```python
import pandas as pd

# (Assumindo que o DataFrame 'df' j√° foi carregado na etapa anterior)

print("\nVerificando os tipos de dados das colunas:")
print(df.info())

print("\nExemplo de valores das primeiras linhas de colunas do tipo 'object' (se houver):")
for col in df.select_dtypes(include='object').columns:
    # Tentaremos imprimir os 3 primeiros valores para ver o conte√∫do
    if len(df[col]) > 0: # Para evitar erro se a coluna estiver vazia
        print(f"Coluna '{col}':")
        # Tentamos pegar os 3 primeiros valores que n√£o s√£o nulos, se existirem
        sample_values = df[col].dropna().head(3).tolist()
        if sample_values:
            print(sample_values)
        else:
            print("  (Nenhum valor n√£o nulo para exibir)")
    else:
        print(f"Coluna '{col}': (Coluna vazia)")
```

```python
colunas_aninhadas = ['customer', 'phone', 'internet', 'account']

for col in colunas_aninhadas:
    if col in df.columns and df[col].apply(lambda x: isinstance(x, dict)).any():
        print(f"Processando coluna '{col}'...")
        # Normaliza a coluna de dicion√°rios em um novo DataFrame tempor√°rio
        df_temp = pd.json_normalize(df[col])

        # Adiciona um prefixo para as novas colunas para evitar conflitos de nomes
        df_temp = df_temp.add_prefix(f'{col}_')

        # Concatena o novo DataFrame tempor√°rio com o DataFrame original
        df = pd.concat([df.drop(columns=[col]), df_temp], axis=1)
        print(f"Coluna '{col}' expandida. Novas colunas adicionadas.")
    else:
        print(f"Coluna '{col}' n√£o encontrada ou n√£o cont√©m dicion√°rios aninhados. Pulando.")


print("\nDataFrame ap√≥s achatar colunas aninhadas:")
print(df.head())
print(f"Novo n√∫mero de linhas e colunas: {df.shape}")
print("\nNovos tipos de dados das colunas:")
print(df.info())
```

```python
print("\nVerificando linhas duplicadas novamente (ap√≥s achatar colunas):")
duplicados = df.duplicated().sum()

if duplicados > 0:
    print(f"N√∫mero de linhas duplicadas encontradas: {duplicados}")
    # Remover linhas duplicadas
    df.drop_duplicates(inplace=True)
    print(f"Linhas duplicadas removidas. Novo n√∫mero de linhas: {df.shape[0]}")
else:
    print("Nenhuma linha duplicada encontrada.")
```

```python
import numpy as np

print("\nVerificando e ajustando tipos de dados:")

print("Tipos de dados atuais:")
print(df.dtypes)
print("\nProcessando a coluna 'account_Charges.Total'...")
df['account_Charges.Total'] = df['account_Charges.Total'].replace(' ', np.nan)
df['account_Charges.Total'] = pd.to_numeric(df['account_Charges.Total'], errors='coerce')
novos_ausentes_total = df['account_Charges.Total'].isnull().sum()
if novos_ausentes_total > 0:
    print(f"ATEN√á√ÉO: Foram encontrados {novos_ausentes_total} valores ausentes em 'account_Charges.Total' ap√≥s a convers√£o.")
    df['account_Charges.Total'].fillna(df['account_Charges.Total'].median(), inplace=True)
    print(f"Valores ausentes em 'account_Charges.Total' preenchidos com a mediana.")
else:
    print("Nenhum novo valor ausente encontrado em 'account_Charges.Total' ap√≥s a convers√£o.")

print("\nVerificando tipos de dados ap√≥s corre√ß√£o de 'account_Charges.Total':")
print(df.dtypes)

print("\nVerificando e padronizando colunas categ√≥ricas:")
for col in df.select_dtypes(include='object').columns:
    if col not in ['customerID']:
        print(f"\nColuna '{col}':")
        df[col] = df[col].astype(str).str.strip().str.lower()
        contagem_valores = df[col].value_counts()
        print("Valores √∫nicos e suas contagens (ap√≥s padroniza√ß√£o):")
        print(contagem_valores)



print("\nVerifica√ß√£o e limpeza de formata√ß√£o conclu√≠das!")
print("\nPrimeiras 5 linhas do DataFrame ap√≥s todas as transforma√ß√µes:")
df.head()
```

```python
dias_no_mes = 30.44
df['Contas_Diarias'] = df['account_Charges.Monthly'] / dias_no_mes
```

```python
import pandas as pd
import numpy as np

colunas_para_binarizar = [
    'Churn',
    'customer_Partner',
    'customer_Dependents',
    'phone_PhoneService',
    'phone_MultipleLines',
    'internet_OnlineSecurity',
    'internet_OnlineBackup',
    'internet_DeviceProtection',
    'internet_TechSupport',
    'internet_StreamingTV',
    'internet_StreamingMovies',
    'account_PaperlessBilling'
]

# Definir o mapeamento
mapeamento_binario = {'yes': 1, 'no': 0}

for col in colunas_para_binarizar:
    if col in df.columns:
        if df[col].astype(str).isin(['yes', 'no']).all():
            df[col] = df[col].map(mapeamento_binario)
        else:
            df[col] = df[col].replace(mapeamento_binario)
df.head()
```

#üìÑRelatorio Final

Este relat√≥rio teve como objetivo analisar o problema de evas√£o de clientes (Churn) na TelecomX para identificar fatores que levam ao cancelamento e gerar insights para reten√ß√£o.

**1. Limpeza e Tratamento de Dados:**
A fase de pr√©-processamento envolveu a importa√ß√£o de dados JSON, o "achatamento" de colunas aninhadas (como informa√ß√µes de cliente, telefone, internet e conta), a verifica√ß√£o de duplicados e o tratamento de valores ausentes/inconsistentes (notavelmente em account_Charges.Total e Churn). As colunas categ√≥ricas foram padronizadas e convertidas para formato bin√°rio (0/1 para 'yes'/'no'), e uma nova feature, Contas_Diarias (faturamento di√°rio), foi criada.

**2. An√°lise Explorat√≥ria de Dados (EDA):**
A an√°lise descritiva inicial revelou a distribui√ß√£o das vari√°veis num√©ricas, como o tempo de contrato (customer_tenure) e os gastos (account_Charges.Monthly, account_Charges.Total). Foi observado um forte desequil√≠brio na vari√°vel Churn, com aproximadamente 73.5% dos clientes permanecendo e 26.5% evadindo.

**As visualiza√ß√µes revelaram insights cruciais:**

Tempo de Contrato (customer_tenure): Clientes que evadem tendem a ter um tempo de perman√™ncia significativamente menor (churn precoce).
Tipo de Contrato (account_Contract): Contratos m√™s a m√™s est√£o associados a uma taxa de Churn drasticamente mais alta do que contratos de um ou dois anos.
Servi√ßos Adicionais: A aus√™ncia de servi√ßos como seguran√ßa online (internet_OnlineSecurity) est√° fortemente ligada a uma maior taxa de Churn. Clientes sem servi√ßos b√°sicos (telefone/internet) parecem ter menor Churn.
Perfil do Cliente: Clientes sem parceiro e sem dependentes mostraram uma maior propens√£o √† evas√£o.
3. Conclus√µes e Recomenda√ß√µes:
Os principais fatores de Churn identificados s√£o: curto tempo de contrato, contratos flex√≠veis (m√™s a m√™s), falta de servi√ßos de seguran√ßa online e clientes sem la√ßos familiares.

**Recomenda√ß√µes:**

Refor√ßar programas de boas-vindas para novos clientes.
Incentivar contratos mais longos com ofertas e benef√≠cios.
Promover ativamente servi√ßos de valor agregado como seguran√ßa online.
Desenvolver ofertas segmentadas para clientes de alto risco identificados (ex: m√™s a m√™s, sem seguran√ßa online, sem parceiro/dependentes).
Analisar a percep√ß√£o de valor/pre√ßo para clientes com faturas mensais elevadas.
Este relat√≥rio fornece uma base de conhecimento para a TelecomX desenvolver estrat√©gias de reten√ß√£o de clientes mais eficazes.
