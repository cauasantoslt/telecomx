```python
import pandas as pd

# (Assumindo que o DataFrame 'df' jÃ¡ foi carregado na etapa anterior)

print("\nVerificando os tipos de dados das colunas:")
print(df.info())

print("\nExemplo de valores das primeiras linhas de colunas do tipo 'object' (se houver):")
for col in df.select_dtypes(include='object').columns:
    # Tentaremos imprimir os 3 primeiros valores para ver o conteÃºdo
    if len(df[col]) > 0: # Para evitar erro se a coluna estiver vazia
        print(f"Coluna '{col}':")
        # Tentamos pegar os 3 primeiros valores que nÃ£o sÃ£o nulos, se existirem
        sample_values = df[col].dropna().head(3).tolist()
        if sample_values:
            print(sample_values)
        else:
            print("  (Nenhum valor nÃ£o nulo para exibir)")
    else:
        print(f"Coluna '{col}': (Coluna vazia)")
```

```python
colunas_aninhadas = ['customer', 'phone', 'internet', 'account']

for col in colunas_aninhadas:
    if col in df.columns and df[col].apply(lambda x: isinstance(x, dict)).any():
        print(f"Processando coluna '{col}'...")
        # Normaliza a coluna de dicionÃ¡rios em um novo DataFrame temporÃ¡rio
        df_temp = pd.json_normalize(df[col])

        # Adiciona um prefixo para as novas colunas para evitar conflitos de nomes
        df_temp = df_temp.add_prefix(f'{col}_')

        # Concatena o novo DataFrame temporÃ¡rio com o DataFrame original
        df = pd.concat([df.drop(columns=[col]), df_temp], axis=1)
        print(f"Coluna '{col}' expandida. Novas colunas adicionadas.")
    else:
        print(f"Coluna '{col}' nÃ£o encontrada ou nÃ£o contÃ©m dicionÃ¡rios aninhados. Pulando.")


print("\nDataFrame apÃ³s achatar colunas aninhadas:")
print(df.head())
print(f"Novo nÃºmero de linhas e colunas: {df.shape}")
print("\nNovos tipos de dados das colunas:")
print(df.info())
```

```python
print("\nVerificando linhas duplicadas novamente (apÃ³s achatar colunas):")
duplicados = df.duplicated().sum()

if duplicados > 0:
    print(f"NÃºmero de linhas duplicadas encontradas: {duplicados}")
    # Remover linhas duplicadas
    df.drop_duplicates(inplace=True)
    print(f"Linhas duplicadas removidas. Novo nÃºmero de linhas: {df.shape[0]}")
else:
    print("Nenhuma linha duplicada encontrada.")
```

```python
import numpy as np

print("\nVerificando e ajustando tipos de dados:")

print("Tipos de dados atuais:")
print(df.dtypes)
print("\nProcessando a coluna 'account_Charges.Total'...")
df['account_Charges.Total'] = df['account_Charges.Total'].replace(' ', np.nan)
df['account_Charges.Total'] = pd.to_numeric(df['account_Charges.Total'], errors='coerce')
novos_ausentes_total = df['account_Charges.Total'].isnull().sum()
if novos_ausentes_total > 0:
    print(f"ATENÃ‡ÃƒO: Foram encontrados {novos_ausentes_total} valores ausentes em 'account_Charges.Total' apÃ³s a conversÃ£o.")
    df['account_Charges.Total'].fillna(df['account_Charges.Total'].median(), inplace=True)
    print(f"Valores ausentes em 'account_Charges.Total' preenchidos com a mediana.")
else:
    print("Nenhum novo valor ausente encontrado em 'account_Charges.Total' apÃ³s a conversÃ£o.")

print("\nVerificando tipos de dados apÃ³s correÃ§Ã£o de 'account_Charges.Total':")
print(df.dtypes)

print("\nVerificando e padronizando colunas categÃ³ricas:")
for col in df.select_dtypes(include='object').columns:
    if col not in ['customerID']:
        print(f"\nColuna '{col}':")
        df[col] = df[col].astype(str).str.strip().str.lower()
        contagem_valores = df[col].value_counts()
        print("Valores Ãºnicos e suas contagens (apÃ³s padronizaÃ§Ã£o):")
        print(contagem_valores)



print("\nVerificaÃ§Ã£o e limpeza de formataÃ§Ã£o concluÃ­das!")
print("\nPrimeiras 5 linhas do DataFrame apÃ³s todas as transformaÃ§Ãµes:")
df.head()
```

```python
dias_no_mes = 30.44
df['Contas_Diarias'] = df['account_Charges.Monthly'] / dias_no_mes
```

```python
import pandas as pd
import numpy as np

colunas_para_binarizar = [
    'Churn',
    'customer_Partner',
    'customer_Dependents',
    'phone_PhoneService',
    'phone_MultipleLines',
    'internet_OnlineSecurity',
    'internet_OnlineBackup',
    'internet_DeviceProtection',
    'internet_TechSupport',
    'internet_StreamingTV',
    'internet_StreamingMovies',
    'account_PaperlessBilling'
]

# Definir o mapeamento
mapeamento_binario = {'yes': 1, 'no': 0}

for col in colunas_para_binarizar:
    if col in df.columns:
        if df[col].astype(str).isin(['yes', 'no']).all():
            df[col] = df[col].map(mapeamento_binario)
        else:
            df[col] = df[col].replace(mapeamento_binario)
df.head()
```

**As visualizaÃ§Ãµes revelaram insights cruciais:**

ServiÃ§os Adicionais: A ausÃªncia de serviÃ§os como seguranÃ§a online (internet_OnlineSecurity) estÃ¡ fortemente ligada a uma maior taxa de Churn. Clientes sem serviÃ§os bÃ¡sicos (telefone/internet) parecem ter menor Churn.
Perfil do Cliente: Clientes sem parceiro e sem dependentes mostraram uma maior propensÃ£o Ã  evasÃ£o.
3. ConclusÃµes e RecomendaÃ§Ãµes:
Os principais fatores de Churn identificados sÃ£o: curto tempo de contrato, contratos flexÃ­veis (mÃªs a mÃªs), falta de serviÃ§os de seguranÃ§a online e clientes sem laÃ§os familiares.
